```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
before = [100, 863, 117, 356, 859, 101]
after = [36, 280, 117, 244, 140, 34]

# Set position of bar on X axis
br1 = np.arange(len(before))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, before, color ='r', width = barWidth,
		edgecolor ='grey', label ='Trước tiền xử lý')
plt.bar(br2, after, color ='g', width = barWidth,
		edgecolor ='grey', label ='Sau tiền xử lý')

# Adding Xticks
plt.xlabel('Nguồn', fontsize = 15)
plt.ylabel('Số bản ghi', fontsize = 15)
plt.xticks([r + 1/2*barWidth for r in range(len(before))],
		['Thê giới di động', 'CellphoneS', 'Viettel Store', 'Tiki', 'Di động Việt', 'Điện máy xanh'])

plt.legend()
plt.show()
```

```python
srcs = ['Thế giới di động', 'CellphoneS', 'Viettel Store',
        'Tiki', 'Di động Việt', 'Điện máy xanh']
 
data = [36, 223, 45, 145, 52, 11]
 
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = srcs, autopct = '%1.2f%%')
 
# show plot
plt.show()
```
