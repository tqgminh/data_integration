```python
import fuzzymatcher
import pandas as pd
```

```python
cps_df = pd.read_csv('clean_data/cellphones.csv')
tgdd_df = pd.read_csv('clean_data/tgdd.csv')
tiki_df = pd.read_csv('clean_data/tiki.csv')
vt_df = pd.read_csv('clean_data/viettel.csv')

cps_df['id'] = cps_df.index
tgdd_df['id'] = tgdd_df.index
tiki_df['id'] = tiki_df.index
vt_df['id'] = vt_df.index

cps_df['src'] = 'cps'
tgdd_df['src'] = 'tgdd'
tiki_df['src'] = 'tiki'
vt_df['src'] = 'vt'
```

```python
def get_brand(s):
    s = s.lower()
    if 'iphone' in s:
        return 'apple'
    elif 'samsung' in s:
        return 'samsung'
    elif 'xiaomi' in s:
        return 'xiaomi'
    elif 'oppo' in s:
        return 'oppo'
    else:
        return 'other'
```

```python
cps_df['thuong_hieu'] = cps_df['ten'].apply(get_brand)
tgdd_df['thuong_hieu'] = tgdd_df['ten'].apply(get_brand)
tiki_df['thuong_hieu'] = tiki_df['ten'].apply(get_brand)
vt_df['thuong_hieu'] = vt_df['ten'].apply(get_brand)
```

```python
apple_cps_df = cps_df[cps_df['thuong_hieu'] == 'apple']
ss_cps_df = cps_df[cps_df['thuong_hieu'] == 'samsung']
xiaomi_cps_df = cps_df[cps_df['thuong_hieu'] == 'xiaomi']
oppo_cps_df = cps_df[cps_df['thuong_hieu'] == 'oppo']
other_cps_df = cps_df[cps_df['thuong_hieu'] == 'other']
```

```python
apple_tgdd_df = tgdd_df[tgdd_df['thuong_hieu'] == 'apple']
ss_tgdd_df = tgdd_df[tgdd_df['thuong_hieu'] == 'samsung']
xiaomi_tgdd_df = tgdd_df[tgdd_df['thuong_hieu'] == 'xiaomi']
oppo_tgdd_df = tgdd_df[tgdd_df['thuong_hieu'] == 'oppo']
other_tgdd_df = tgdd_df[tgdd_df['thuong_hieu'] == 'other']
```

```python
apple_tiki_df = tiki_df[tiki_df['thuong_hieu'] == 'apple']
ss_tiki_df = tiki_df[tiki_df['thuong_hieu'] == 'samsung']
xiaomi_tiki_df = tiki_df[tiki_df['thuong_hieu'] == 'xiaomi']
oppo_tiki_df = tiki_df[tiki_df['thuong_hieu'] == 'oppo']
other_tiki_df = tiki_df[tiki_df['thuong_hieu'] == 'other']
```

```python
apple_vt_df = vt_df[vt_df['thuong_hieu'] == 'apple']
ss_vt_df = vt_df[vt_df['thuong_hieu'] == 'samsung']
xiaomi_vt_df = vt_df[vt_df['thuong_hieu'] == 'xiaomi']
oppo_vt_df = vt_df[vt_df['thuong_hieu'] == 'oppo']
other_vt_df = vt_df[vt_df['thuong_hieu'] == 'other']
```

```python
left_on = ["ten", "gia",  "ram", "bo_nho_trong", "kich_thuoc_man_hinh"]
right_on = ["ten", "gia",  "ram", "bo_nho_trong", "kich_thuoc_man_hinh"]
```

# Matching data Apple

## TGDD vs CellphoneS

```python
apple_matcher = fuzzymatcher.link_table(apple_cps_df, apple_tgdd_df, left_on, right_on, left_id_col = "id", right_id_col = "id")
apple_matcher = apple_matcher[apple_matcher['match_score'] > 0.15]
apple_matcher
```

```python
values = apple_matcher['__id_right'].tolist()
apple_tgdd_df = apple_tgdd_df.query("id not in @values")

apple_df = pd.concat([apple_cps_df, apple_tgdd_df])
apple_df = apple_df.reset_index(drop=True)
apple_df['id'] = apple_df.index
print(len(apple_df))
```

## Viettel vs CellphoneS

```python
apple_matcher = fuzzymatcher.link_table(apple_df, apple_vt_df, left_on, right_on, left_id_col = "id", right_id_col = "id")
apple_matcher = apple_matcher[apple_matcher['match_score'] > 0.16]
apple_matcher
```

```python
values = apple_matcher['__id_right'].tolist()
apple_vt_df = apple_vt_df.query("id not in @values")

apple_df = pd.concat([apple_df, apple_vt_df])
apple_df = apple_df.reset_index(drop=True)
apple_df['id'] = apple_df.index
print(len(apple_df))
```

## Tiki vs CellphoneS

```python
apple_matcher = fuzzymatcher.link_table(apple_df, apple_tiki_df, left_on, right_on, left_id_col = "id", right_id_col = "id")
apple_matcher = apple_matcher[apple_matcher['match_score'] > 0.1]
apple_matcher
```
